```python
from cxai import factory
from cxai import utils as putils
```

```python
img = putils.load_image("../tests/data/castle.jpg")
label = 483
```

```python
putils.viz.imshow(img)
```

```python
model, (rc_transform, input_transform) = factory.make_model("torchvision-vgg16-imagenet")
```

```python
explainer = factory.make_explainer("lrp", model)
```

```python
logits, heatmap = explainer.explain(
    input_transform(img), label
)
heatmap.shape
```

```python
putils.viz.heatmap(heatmap.sum(axis=0))
```
